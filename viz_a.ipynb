```
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```
titanic = sns.load_dataset('titanic')
titanic.head()
```

# Did age impact chances of survival on the titanic?

```
age_sum = titanic.groupby('age', as_index=False).sum()
sns.lmplot(x='age', y='survived', data=age_sum)
print()
```

# Did the passenger's class correlate to their chances of survival? 

```
fare_total = titanic.groupby('age', as_index=False).sum()
sns.lmplot(x='pclass', y='survived', data=fare_total)
print()
```

# Did the amount of money a passenger paid correlate with passenger class?

```
fare_total = titanic.groupby('age', as_index=False).sum()
sns.swarmplot(x='fare', y='pclass', data=fare_total)
# sns.set_xticklabels(labels=titanic['fare'], rotation=90)
print()
```

# Did gender and passenger class raise or lower chances of survival?

```
sns.pointplot(x="class", y="survived", hue="sex", data=titanic, palette={"male":"b", "female":"m"}, markers=["^","o"], linestyles=[":","-."])
print()
```
